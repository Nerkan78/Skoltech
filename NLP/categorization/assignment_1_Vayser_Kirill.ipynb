## 2.1 Methodology 

### 2.1.1. First attempt. Replication and improvements of seminar.
This is one of my first serious attempt to solve NLP task. so I decided to start with seminar and introduce some improvements.
1. Preprocessing

  Preprocessing included some standart techniques: lemmatization, stemming, removal of punctuation and stop words.

2. Processing 

  For embedding construction I used fasttext model.

3. Model

I splitted multitarget multiclass problem in several one target multiclass problems and used LSTM cells for each of them. I've experimented with number of layers and hidden sizes of LSTM but failed to achieve significant improvement.

In this approach I obtained quality below baseline, so I looked elsewhere. 


### 2.1.2 Second attempt. Manual usage of transformers

One of the modern architectures for NLP problems is BERT, so I decided to try it. In description of BERT it was said that it adds special CLS token for classification. I decided to exploit this idea to avoid multitarget problem. Instead  I added to sentences words which indicates the target, so the problem became single target multiclass by the price of expanded dataset. However the results were very poor and it seemed that model didn't train. I failed to discover problem in time. Later it appeared that most likely inserted indicators were truncated, so model just was receiveing idenctial inputs with different labels.

### 2.3.2 Final attempt. Transformers out of the box. 

To find solution to abovementioned problem I went to the Internet with that question. I found library which allowed to use transformers out of the box and thereby exclude possibility of mistake in implementation. Also, I've read documentation of transformers more carefully and placed indicators in the beginning of sentences. I used AlBert model and it gave me best result. Unfortunately colab failed during training so the undertrained model was used. Luckily, it beated the baseline.

### 3.3.2 Preprocessing

```python
import pandas as pd
train_df = pd.read_csv("train.tsv", sep="\t")
val_df = pd.read_csv("val_empty.tsv", sep="\t")
test_df = pd.read_csv("test-no_labels.tsv", sep="\t")
```

```python
import numpy as np
LABEL_COLUMNS = ['masks_stance',	'masks_argument',	'quarantine_stance',	'quarantine_argument',	'vaccines_stance',	'vaccines_argument']
statements = ['«Вакцинация полезна для общества»', '«Введение и соблюдение карантина полезно для общества»', '«Ношение масок полезно для общества»'][::-1]

full_y_train = train_df[LABEL_COLUMNS].values.flatten() + 1

full_X_train = []
X_test = []

for text in train_df.text.values:
  for statement in statements:
    full_X_train.append('«Позиция»' + '/' + statement + '/' + text)
    full_X_train.append('«Довод»' + '/' + statement + '/' + text)
full_X_train = np.vstack(full_X_train)

for text in test_df.text.values:
  for statement in statements:
    X_test.append('«Позиция»' + '/' + statement + '/' + text)
    X_test.append('«Довод»' + '/' + statement + '/' + text)
X_test = np.vstack(X_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

X_train, X_val, y_train, y_val = train_test_split(full_X_train, full_y_train, shuffle=False, test_size=0.1)
```

```python
train=pd.DataFrame({'text': list(X_train.reshape(-1,)), 'labels': y_train})
val=pd.DataFrame({'text': list(X_val.reshape(-1,)), 'labels': y_val})
```
